```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

print("All imports successful")
df=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
print(df.head())
df_new=df.drop(columns=["PassengerId","Name","Ticket","Cabin"])
print(df_new)
missing_values=df_new.isnull().sum()
print(missing_values)
df_new=df_new.dropna(subset=["Embarked"])
Median_Age=df_new["Age"].median()
df_new["Age"] = df_new["Age"].fillna(Median_Age)
print(df_new)
check=df_new.isnull().sum()

print(check)
df_new=df_new.copy()
df_new["Sex"]=df_new["Sex"].map({"male":0,"female":1})
df_new["Embarked"]=df_new["Embarked"].map({"S":0,"C":1,"Q":2})
print(df_new.head())
```

```python
x=df_new.drop(columns=["Survived"])
y=df_new["Survived"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
accuracy_percentage=accuracy*100
print(f"Model accuracy: {accuracy_percentage:.2f}%")
```
